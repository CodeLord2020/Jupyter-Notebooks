```python
#SETTING DATES, BANKS AND TICKERS

start_date = dt.datetime(2006, 1, 1)
end_date = dt.datetime(2022, 12, 31)
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']

stocks = ['Bank of America', 'CitiGroup', 'Goldman Sachs', 'JPMorgan', 'Morgan Stanley', 'Wells Fargo&Co']

#Using yfinace cos GOOGLE and YAHOO with DataReader seems to b discontinued or not working for now.
```

```python
#Bank of America stock data

BAC = yf.download("BAC", start_date, end_date)
#BAC['Day'] = BAC['Date'].date()
BAC['Date'] = BAC.index
BAC = BAC[['Open', 'Low', 'High', 'Close', 'Adj Close', 'Volume']]
#BAC.reset_index(drop = True, inplace = True)
BAC.head()
```

```python
#CitiGroup stock data


C = yf.download("C", start_date, end_date)
C['Date'] = C.index
C = C[['Open', 'Low', 'High', 'Close', 'Adj Close', 'Volume']]
#C.reset_index(drop = True, inplace = True)
C.head()
```

```python
#Goldman Sachs stock data


GS = yf.download("GS", start_date, end_date)
GS['Date'] = GS.index
GS = GS[[ 'Open', 'Low', 'High', 'Close', 'Adj Close', 'Volume']]
#GS.reset_index(drop = True, inplace = True)
GS.head()
```

```python
#JPMorgan stock data
JPM = yf.download("JPM", start_date, end_date)
JPM['Date'] = JPM.index
JPM = JPM[['Open', 'Low', 'High', 'Close', 'Adj Close', 'Volume']]
#JPM.reset_index(drop = True, inplace = True)
JPM.head()
```

```python
#Morgan Stanley
MS = yf.download("MS", start_date, end_date)
MS['Date'] = MS.index
MS = MS[[ 'Open', 'Low', 'High', 'Close', 'Adj Close', 'Volume']]
#MS.reset_index(drop = True, inplace = True)
MS.head()
```

```python
#Wells Fargo&Co stock data


WFC = yf.download("WFC", start_date, end_date)
WFC['Date'] = WFC.index
WFC = WFC[['Open', 'Low', 'High', 'Close', 'Adj Close', 'Volume']]
#WFC.reset_index(drop = True, inplace = True)
WFC.tail()
```

```python
#CONCATINATING BANKS STOCK DATA

bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC], axis = 1, keys = tickers)
pd.set_option('display.max_columns',None)
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
stocks = ['Bank of America', 'CitiGroup', 'Goldman Sachs', 'JPMorgan', 'Morgan Stanley', 'Wells Fargo&Co']
bank_stocks.index = bank_stocks.index.date
bank_stocks.head()
```

```python
#REMOVING DUPLICATES (Unlikely to b one though)


bank_stocks.drop_duplicates( inplace = True )
bank_stocks.describe()
```

```python
#EXPORTING TO WORK OFFLINE

bank_stocks.to_csv("bank_stocks.csv")
bank_stocks.to_excel("bank stocks.xlsx", sheet_name='Sheet_name_1')
bank_stocks.info()
```

# MACHINE LEARNING

Using Goldman Sachs(open,close, low, high,close) stock prices to predict  'Morgan Stanley closing stock prices

There is a high positive correlation of 0.86 between their stock prices

```python
X = GS[['Open','Low','High', 'Close']]
y = MS['Close']

from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyRegressor
from sklearn.metrics import accuracy_score
#np.array(X)
#X.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,  random_state=101, test_size = 0.3)
clf = LazyRegressor(verbose = 0, ignore_warnings = True)
#X_train, X_test, y_train, y_test = train_test_split(x , y, test_size=.3, random_state =123)
model_predictions = clf.fit(x_train, x_test, y_train, y_test)
```

```python
print(model_predictions)
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import Lars
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression


from sklearn.metrics import accuracy_score

reg2 = Lars()
reg3 = GradientBoostingRegressor()
reg4 = LinearRegression()

reg = VotingRegressor(estimators = [ ('lar', reg2), ('gbr', reg3), ('lr', reg4)] )
```

```python


reg.fit(x_train, y_train)
pred = reg.predict(x_test)
#data = pd.DataFrame(y_test).concat(pd.DataFrame(pred, axis = 1))
#result = pd.DataFrame(y_test,pred, columns =['Predicted Y', 'Exact_Y'])
print(reg.score(x_test, y_test))
```
